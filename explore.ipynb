```python
# Base URL to your hosted parquet files on Hugging Face
base_url = "https://huggingface.co/datasets/ChernovAndrei/reco-fm-data/resolve/main"

# List of parquet file names to load
files = {
    "train": "train_data.parquet",
    "val": "val_data.parquet",
    "test": "test_data.parquet",
    "title_embeddings": "title_embeddings.parquet"
}

# Load each dataset individually
datasets = {}
for split_name, filename in files.items():
    print(f"Loading {split_name} from {filename}...")
    datasets[split_name] = load_dataset(
        path="parquet",
        data_files=f"{base_url}/{filename}",
        split=None
    )
    # print(f"{split_name} loaded. Rows: {len(datasets[split_name])}")
```

```python
# Create mapping dictionaries for user_id and parent_asin
user_map = {user_id: idx+1 for idx, user_id in enumerate(list(np.unique(datasets['train']['train']['user_id'] + datasets['val']['train']['user_id'] + datasets['test']['train']['user_id'])))}
item_map = {parent_asin: idx+1 for idx, parent_asin in enumerate(list(np.unique(datasets['train']['train']['parent_asin'] + datasets['val']['train']['parent_asin'] + datasets['test']['train']['parent_asin'])))}

# # Create new dataframe with mapped IDs
# mapped_df = datasets['train']['train'].to_pandas().copy()
# mapped_df['uid'] = mapped_df['user_id'].map(user_map)
# mapped_df['sid'] = mapped_df['parent_asin'].map(item_map)

# # Print some statistics
# print(f"Number of unique users: {len(user_map)}")
# print(f"Number of unique items: {len(item_map)}")
# print("\nFirst few rows of the mapped dataframe:")
# print(mapped_df[['user_id', 'uid', 'parent_asin', 'sid']].head())
```

```python
df_val = datasets['val']['train'].to_pandas()
df_test = datasets['test']['train'].to_pandas()

df_val.head()
df_test.head()
```

```python
# Map user_id and parent_asin to uid and sid
df_val_mapped = df_val.copy()
df_val_mapped['uid'] = df_val_mapped['user_id'].map(user_map)
df_val_mapped['sid'] = df_val_mapped['parent_asin'].map(item_map)

# Map history items to sid
df_val_mapped['history'] = df_val_mapped['history'].apply(lambda x: ' '.join([str(item_map[item]) for item in x.split()]) if isinstance(x, str) else '')

# Find rows where sid is NaN
nan_sids = df_val_mapped[df_val_mapped['sid'].isna()]
print("\nRows where sid is NaN:")
print(nan_sids[['user_id', 'parent_asin', 'sid']])
```

```python
df_val_mapped.sort_values(by='uid')
```

```python
final_train = {row['uid']: [int(x) for x in row['history'].split()] if row['history'] else [] for _, row in df_val_mapped.iterrows()}
```

```python
# Sort final_train by keys
final_train = dict(sorted(final_train.items()))

# Create final_val dictionary with uid->sid mapping and sort by keys
final_val = {row['uid']: [row['sid']] for _, row in df_val_mapped.iterrows()}
final_val = dict(sorted(final_val.items()))
```

```python
df_test_mapped = df_test.copy()
df_test_mapped['uid'] = df_test_mapped['user_id'].map(user_map)
df_test_mapped['sid'] = df_test_mapped['parent_asin'].map(item_map)

df_test_mapped.head()

final_test = {row['uid']: [row['sid']] for _, row in df_test_mapped.iterrows()}
final_test = dict(sorted(final_test.items()))



```

```python
meta = datasets['title_embeddings']['train'].to_pandas().drop(['embedding'], axis=1)
meta['sid'] = meta['parent_asin'].map(item_map)
meta.head()
```

```python
final_meta = {row['sid']: row['title'] for _, row in meta.iterrows()}
final_meta = dict(sorted(final_meta.items()))

final_meta
```

```python
final_dataset = {}
final_dataset['train'] = final_train
final_dataset['val'] = final_val
final_dataset['test'] = final_test
final_dataset['meta'] = final_meta
final_dataset['smap'] = item_map
final_dataset['umap'] = user_map
final_dataset
```
